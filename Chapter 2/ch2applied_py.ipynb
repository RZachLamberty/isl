#### 8.c.ii

```python
# Special pandas caveat: whereas regular old college[, 1:10] or college[1:10]
# would work in R, in pandas you need to use the index locator (.iloc) method,
# or you need to use the column names themselves 

# I also took the liberty of making this a bit prettier; lowering the alpha 
# layer, replacing the histograms with kernel densities, and increasing the
# size of the figure to accomodate the huge number of figures
pd.tools.plotting.scatter_matrix(
    college.iloc[:, :11],
    alpha=0.2,
    diagonal='kde',
    figsize=(8, 8)
);
```

#### 8.c.iii

We first need to make `Private` a factor (it is just a Yes v No string)

```python
# in R it is "as.factor()", in pandas you have to cast it as a category
college.loc[:, 'Private'] = college.Private.astype('category')
```

```python
# R's default plot takes care of the side-by-side plots nicely;
# in pandas we need to do a bit more work
college[['Private', 'Outstate']].boxplot('Outstate', by='Private')
#.plot(kind='box')
#college.boxplot?
#college.head()
```
