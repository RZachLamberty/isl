#### 8.c.iii

We first need to make `Private` a factor (it is just a Yes v No string)

```python
# in R it is "as.factor()", in pandas you have to cast it as a category
college.loc[:, 'Private'] = college.Private.astype('category')
```

```python
# R's default plot takes care of the side-by-side plots nicely;
# in pandas we need to do a bit more work (but not much)
college[['Private', 'Outstate']].boxplot('Outstate', by='Private');
```

#### 8.c.iv

We wish to create a qualitative variable called Elite by binning the `Top10perc` variable.

```python
college.loc[:, 'Elite'] = "No"
college.loc[college.Top10perc > 50, 'Elite'] = "Yes"
college.loc[:, 'Elite'] = college.loc[:, 'Elite'].astype('category') 
college.Elite.describe()
```

```python
college[['Elite', 'Outstate']].boxplot('Outstate', by='Elite');
```

```python
def pdrange(row):
    return pd.Series({'min': row.min(), 'max': row.max()}, index=['min', 'max'])
```
